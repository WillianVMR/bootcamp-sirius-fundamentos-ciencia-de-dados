```python
# Importação da Base de Dados

import pandas as pd

netflix = pd.read_csv('titles.csv')
```

```python
# Encontrar Medidas nos Dados com describe()

netflix[['imdb_score', 'tmdb_popularity']].describe().round(2)
```

```python
# Calcular Medidas com GroupBy

netflix.groupby('production_countries')[['tmdb_popularity', 'imdb_score']].mean().round(2)

```

```python
netflix.groupby('production_countries')[['tmdb_popularity', 'imdb_score']].median().round(2)
```

```python
netflix.groupby('production_countries')[['tmdb_popularity', 'imdb_score']].var().round(2)
```

```python
netflix.groupby('production_countries')[['tmdb_popularity', 'imdb_score']].std().round(2)
```

```python
netflix.groupby('production_countries')[['tmdb_popularity', 'imdb_score']].quantile(0.10).round(2)
```

```python
netflix.groupby('production_countries')[['tmdb_popularity', 'imdb_score']].quantile(0.25).round(2)
```

```python
# BoxPlots

import seaborn as sns

sns.boxplot(y='imdb_score', x = 'production_countries', data=netflix)
```
